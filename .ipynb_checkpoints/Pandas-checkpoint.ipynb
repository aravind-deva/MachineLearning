```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston=load_boston()

boston.data.shape

df=pd.DataFrame(boston.data,columns=boston.feature_names)

print(df.head(5))

index= np.arange (30,536)

df=pd.DataFrame(boston.data,columns=boston.feature_names,index=index)

df.head(5)
```

```python
# select  a row. It is a pandas series object with key(cols)  as columns
df.loc[30]
print(type(df.loc[30]))

# select a column, it is a pandas series object with keys as row_indexs
df['CRIM']
print(type(df['CRIM']))

print(df['CRIM'][31])
```

```python
print('Selecting a column gives a Series with row label as index/columns of series,name as column name')

df['CRIM']

```

```python
print('selecting 2 columns returns a data frame')
df[['CRIM','ZN']]
```

```python
print('Selecting a row gives a Series with columns as index/columns of series, name as row label')
df.loc[30]
```

```python
df.sort_values(['CRIM','ZN'],ascending=[True,True])
```
